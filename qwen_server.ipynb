```python
import requests

from llamavision import filepath_to_base64

filepath = "imgs/impulse.png"
print(filepath_to_base64(filepath)[:10], "...", filepath_to_base64(filepath)[-10:])
```

```python
response_format_graph = {
      "type": "json_schema",
      "json_schema": {
        "name": "parse_graph_response",
        "strict": "true",
        "schema": {
          "type": "object",
          "properties": {
            "x-tickmarks": {
              "type": "string",
              "description": "The tickmarks on the x-axis"
            },
            "y-tickmarks": {
              "type": "string",
              "description": "The tickmarks on the y-axis"
            },
            "image-summary": {
              "type": "string",
              "description": "A summary of the image generated by the assistant"
            }
          },
        "required": ["x-tickmarks", "y-tickmarks", "image-summary"]
        }
      }
    }
```

```python

headers = {
    'Content-Type': 'application/json',
}
image_path = 'imgs/impulse.png'
image_url = filepath_to_base64(image_path)

json_data = {
    'model': 'mlx-community/Qwen2-VL-7B-Instruct-8bit',
    'messages': [
        {
            'role': 'user',
            'content': [
                {
                    'type': 'text',
                    'text': 'Extract information from the image and return it as a JSON object.',
                },
                {
                    'type': 'image_url',
                    'image_url': {
                        'url': f'data:image/png;base64,{image_url}',
                    },
                },
            ],
        },
    ],
    'response_format': response_format_graph,
    'temperature': 0.7,
    'max_tokens': -1,
    'stream': False,
}
```

```python
response = requests.post('http://127.0.0.1:1234/v1/chat/completions', headers=headers, json=json_data)
response_data = response.json()
for item in response_data:
    print(item, response_data[item])
```
